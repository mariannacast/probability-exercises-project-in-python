## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np 

import random


def throw_dice():
    return np.random.randint(1,6) 

def simulation(n):
    count = 0
    for _ in range(n):
        throw = throw_dice() + throw_dice () 
        if throw > 7 or throw % 2 == 0: 
            count += 1
    return count/n*100

print(simulation(100))
       

```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import numpy as np

# Crear la caja de bolas
ball_box = {}

for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif 10 <= i < 30:
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Definir el número de simulaciones
num_simulations = 1000

# Convertir las bolas en una lista para seleccionar aleatoriamente
ball_list = list(ball_box.values())

# Realizar la simulación de tomar 5 bolas con reemplazo
resultados = [np.random.choice(ball_list, 5, replace=True) for _ in range(num_simulations)]

# Contar los casos donde obtenemos 3 blancas y 2 rojas
condicion_1 = [np.sum(sim == 'White') == 3 and np.sum(sim == 'Red') == 2 for sim in resultados]

# Contar los casos donde todas las bolas sean del mismo color
condicion_2 = [len(set(sim)) == 1 for sim in resultados]

# Calcular las probabilidades
prob_3_blancas_2_rojas = np.sum(condicion_1) / num_simulations
prob_mismo_color = np.sum(condicion_2) / num_simulations

print(f"Probabilidad de 3 blancas y 2 rojas: {prob_3_blancas_2_rojas}")
print(f"Probabilidad de todas del mismo color: {prob_mismo_color}")
```
